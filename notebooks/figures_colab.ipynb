# Causal-VITGN — Generate all 7 paper figures

This notebook regenerates every figure used in the paper from the trained checkpoint.

**Prereqs:**
1. The `causal-vitgn` repo is cloned into Colab at `/content/causal-vitgn`.
2. `best_model.pt` lives at `/content/causal-vitgn/checkpoints/best_model.pt`.
3. `data/processed/{train,val,test}.pkl` are present.

**Output:** Each figure is saved as a PDF (for LaTeX) and a PNG (for slides) under `figures/`.

```python
# Restore the trained checkpoint + processed data from Drive if missing
import os, shutil

DRIVE_SRC = '/content/drive/MyDrive/causal-vitgn-ckpt'
os.makedirs('checkpoints', exist_ok=True)
os.makedirs('data/processed', exist_ok=True)

ckpt = 'checkpoints/best_model.pt'
if not os.path.exists(ckpt):
    shutil.copy(f'{DRIVE_SRC}/best_model.pt', ckpt)
    print(f'restored {ckpt}')

if not os.path.exists('data/processed/metadata.json'):
    for f in os.listdir(f'{DRIVE_SRC}/data_processed'):
        shutil.copy(f'{DRIVE_SRC}/data_processed/{f}', f'data/processed/{f}')
    print('restored data/processed')

!ls -lh checkpoints/ data/processed/
```

```python
# Save the figures back to Drive so you can download them for LaTeX
import shutil
DRIVE_DST = '/content/drive/MyDrive/causal-vitgn-ckpt/figures'
os.makedirs(DRIVE_DST, exist_ok=True)
for f in os.listdir('figures'):
    shutil.copy(f'figures/{f}', f'{DRIVE_DST}/{f}')
print(f'Copied {len(os.listdir("figures"))} files to {DRIVE_DST}')
!ls -lh figures/
```
